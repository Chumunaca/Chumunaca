AHORA VAMOS A MODELAR

```python
# Definir target y predictoras
X = df[['laterality', 'view_position', 'breast_density', 'finding_categories', 'Age']]
y = df['breast_birads']
```

```python
# Dividir en train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
# Preprocesamiento
num_features = ['Age']
cat_features = ['laterality', 'view_position', 'breast_density', 'finding_categories']

num_pipeline = Pipeline([
    # Si quieres imputar de nuevo: ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('cat', cat_pipeline, cat_features)
])
```

```python
# Pipeline completo con SMOTE
model_pipeline = ImbPipeline([
    ('preproc', preprocessor),
    ('smote', SMOTE(random_state=42)),
    # Aqu√≠ ir√≠a el estimador, ej.:
    # ('clf', LogisticRegression(multi_class='multinomial', max_iter=1000))
])
```

```python
#Ajustar solo el preprocesamiento y el muestreo para inspecci√≥n
X_res, y_res = model_pipeline.fit_resample(X_train, y_train)
print("Distribuci√≥n original en train:\n", y_train.value_counts())
print("\nDistribuci√≥n tras SMOTE:\n", pd.Series(y_res).value_counts())
```

```python
# Visualizaci√≥n de la distribuci√≥n de 'breast_birads' despu√©s de aplicar SMOTE
plt.figure(figsize=(8, 5))
pd.Series(y_res).value_counts().sort_index().plot(kind='bar', color='orange')
plt.title('Distribuci√≥n de BI-RADS despu√©s de SMOTE')
plt.xlabel('Categor√≠a BI-RADS')
plt.ylabel('N√∫mero de casos (sint√©ticos + originales)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

REGRESI√ìN LOGISTICA

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt

# Pipeline con SMOTE + preprocesador + Regresi√≥n Log√≠stica
pipe_lr = Pipeline([
    ('preproc', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('clf', LogisticRegression(multi_class='multinomial', max_iter=1000, random_state=42))
])

# Entrenar y predecir
pipe_lr.fit(X_train, y_train)
y_pred = pipe_lr.predict(X_test)

# Reporte
print(classification_report(y_test, y_pred, zero_division=0))

# Matriz de confusi√≥n
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(4,3))
plt.imshow(cm, cmap=plt.cm.Blues)
plt.title('Confusion Matrix ‚Äî Regresi√≥n Log√≠stica')
plt.colorbar()
classes = np.unique(y_test)
ticks = np.arange(len(classes))
plt.xticks(ticks, classes, rotation=45)
plt.yticks(ticks, classes)
plt.xlabel('Predicha')
plt.ylabel('Verdadera')
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

pipe_rf = Pipeline([
    ('preproc', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('clf', RandomForestClassifier(n_estimators=100, random_state=42))
])

pipe_rf.fit(X_train, y_train)
y_pred = pipe_rf.predict(X_test)

print(classification_report(y_test, y_pred, zero_division=0))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(4,3))
plt.imshow(cm, cmap=plt.cm.Blues)
plt.title('Confusion Matrix ‚Äî Random Forest')
plt.colorbar()
plt.xticks(ticks, classes, rotation=45)
plt.yticks(ticks, classes)
plt.xlabel('Predicha')
plt.ylabel('Verdadera')
plt.tight_layout()
plt.show()
```

```python
from sklearn.svm import SVC

pipe_svc = Pipeline([
    ('preproc', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('clf', SVC(kernel='rbf', probability=True, random_state=42))
])

pipe_svc.fit(X_train, y_train)
y_pred = pipe_svc.predict(X_test)

print(classification_report(y_test, y_pred, zero_division=0))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(4,3))
plt.imshow(cm, cmap=plt.cm.Blues)
plt.title('Confusion Matrix ‚Äî SVM')
plt.colorbar()
plt.xticks(ticks, classes, rotation=45)
plt.yticks(ticks, classes)
plt.xlabel('Predicha')
plt.ylabel('Verdadera')
plt.tight_layout()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

pipe_dt = Pipeline([
    ('preproc', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('clf', DecisionTreeClassifier(random_state=42))
])

pipe_dt.fit(X_train, y_train)
y_pred = pipe_dt.predict(X_test)

print(classification_report(y_test, y_pred, zero_division=0))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(4,3))
plt.imshow(cm, cmap=plt.cm.Blues)
plt.title('Confusion Matrix ‚Äî √Årbol de Decisi√≥n')
plt.colorbar()
plt.xticks(ticks, classes, rotation=45)
plt.yticks(ticks, classes)
plt.xlabel('Predicha')
plt.ylabel('Verdadera')
plt.tight_layout()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

pipe_knn = Pipeline([
    ('preproc', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('clf', KNeighborsClassifier(n_neighbors=5))
])

pipe_knn.fit(X_train, y_train)
y_pred = pipe_knn.predict(X_test)

print(classification_report(y_test, y_pred, zero_division=0))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(4,3))
plt.imshow(cm, cmap=plt.cm.Blues)
plt.title('Confusion Matrix ‚Äî KNN')
plt.colorbar()
plt.xticks(ticks, classes, rotation=45)
plt.yticks(ticks, classes)
plt.xlabel('Predicha')
plt.ylabel('Verdadera')
plt.tight_layout()
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier

pipe_mlp = Pipeline([
    ('preproc', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('clf', MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, random_state=42))
])

pipe_mlp.fit(X_train, y_train)
y_pred = pipe_mlp.predict(X_test)

print(classification_report(y_test, y_pred, zero_division=0))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(4,3))
plt.imshow(cm, cmap=plt.cm.Blues)
plt.title('Confusion Matrix ‚Äî Red Neuronal')
plt.colorbar()
plt.xticks(ticks, classes, rotation=45)
plt.yticks(ticks, classes)
plt.xlabel('Predicha')
plt.ylabel('Verdadera')
plt.tight_layout()
plt.show()
```

```python
# 6. Definir modelos
models = {
    'Regresi√≥n Log√≠stica': LogisticRegression(multi_class='multinomial', max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'SVM': SVC(kernel='rbf', probability=True, random_state=42),
    '√Årbol de Decisi√≥n': DecisionTreeClassifier(random_state=42),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'Red Neuronal': MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, random_state=42)
}

# 7. Entrenar y evaluar
for name, clf in models.items():
    print(f"\n===== MODELO: {name} =====")

    # Pipeline completo con SMOTE en train
    pipeline = ImbPipeline([
        ('preproc', preprocessor),
        ('smote', SMOTE(random_state=42)),
        ('clf', clf)
    ])

    # Entrenamiento
    pipeline.fit(X_train, y_train)

    # Predicci√≥n
    y_pred = pipeline.predict(X_test)

    # M√©tricas
    print("Classification Report:")
    print(classification_report(y_test, y_pred, zero_division=0))

    # Matriz de confusi√≥n
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:\n", cm)

    # Gr√°fico de matriz de confusi√≥n
    plt.figure(figsize=(5,4))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(f'Matriz de Confusi√≥n - {name}')
    plt.colorbar()
    classes = np.unique(y)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    plt.ylabel('Etiqueta verdadera')
    plt.xlabel('Etiqueta predicha')
    plt.tight_layout()
    plt.show()
```
